## Краткий обзор рекомендательных систем

### Определение
__Рекомендательные системы__ являются одним из инстурментов в бизнесе, которые позволяют сориентирвовать клиента в большом разнообразии товаров.

_Типичный пользователь в онлайн магазине H&M_

<img src='https://thumbs.gfycat.com/SmoothJauntyBighornedsheep-max-1mb.gif' alt='Типичный пользователь в онлайн магазине H&M' />

### Виды
Рекомендательные системы по принципу работы делятся на:
1. Рейтинговые;
2. На основе содержания;
3. Коллаборативные.
---
#### Рейтинговые
Алгоритм заключается в рекомендации самых популярных товаров клиенту.

##### Плюсы:
1. Простая реализация;
2. Отсутсвие проблемы холодного старта*.

##### Минусы:
1. Не учитываются персональные предпочтения пользователей.
---
#### На основе содержания
Алгоритм основывается на предпочтениях клиента (истории его покупок). Система показывает те товары, категориями которых он интересовался.

##### Плюсы:
1. Персонализированная рекомендация.

##### Минусы:
1. Рекомендация не показывает новый контент для пользователя;
2. Проблема холодного старта.
---
#### Коллаборативные
Алгоритм основывается на предпочтениях группы на основе клиентов/предметов рекомендации.

##### Плюсы:
1. Рекомендации могут показывать новый контент:
2. Простоая реализация.

##### Минусы:
1. Модели могут долго обучатся и много занимают место;
2. Проблема холодного старта.
---
*__Проблема холодного старта__ возникает, когда данных о пользователе/товаров еще мало. Такое возникает, когда пользоваль впервые воспользовался услугой или сервис только начал свою работу. Такое может решаться ручной корректировкой алгоритма, использованием рейтинговой рекомендательной системой, начальным определением предпочтений пользователя (в случае нового пользователя).
